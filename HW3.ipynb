Central tendency of Philly park size

```python
meanps = ppark['ACREAGE'].mean()
print("mean:", meanps)
```

```python
medianps = ppark['ACREAGE'].mean()
print("median:", medianps)
```

what can you conclude from the values of the mean and median you extracted?

The median and mean are equal together, so the distribution is symmetric.

```python
ax = ppark.ACREAGE.hist(bins=35)
ax.set_xlabel('Size of Philly park [ac]')
ax.axvline(meanps, color='red', linestyle='solid', linewidth=1.6, alpha = .8)
ax.axvline(medianps, color='green', linestyle='solid', linewidth=1.6, alpha = .8)
ax.legend([meanps, medianps], labels = ['mean', 'median'])
```

Figure 1: This figures shows the distribution of Philly's park sizes in acres. As it is shown, the median and mean values are same showing the distribution is symetric. Based on this figure, most of the parks are less than 125 acres and; however majority of them are near 0 acres, there are some few parks more than 250 and 1000 acres. 

```python
stdps = ppark['ACREAGE'].std()
print('Standard deviation:', stdps)
```

Parks smaller than 10 acres:

```python
smallps = ppark.loc[ppark['ACREAGE'] < 10]
smallps.head()
```

```python
meanps = smallps['ACREAGE'].mean()
medianps = smallps['ACREAGE'].median()
print('Mean value:', meanps)
print('Median value:', medianps)
```

```python
ax = smallps.ACREAGE.hist(bins=35)
ax.set_xlabel('Size of Philly park [ac]')
ax.axvline(meanps, color='red', linestyle='solid', linewidth=1.6, alpha = .8)
ax.axvline(medianps, color='green', linestyle='solid', linewidth=1.6, alpha = .8)
ax.legend([meanps, medianps], labels = ['mean', 'median'])
```

Figure 2: The figure shows the distribution of Philadelphia's park, having an area of fewer than 10 acres. The mean value is greater than the median, so the distribution has a positive skew. The majority of parks have areas between 0 and 2 acres. This figure shows that by eliminating large parks, the mean and median are much closer to the middle of the chart as the outliers don't influence results.

The challenge is to visualize the sizes of the parent parks.

```python
park_group = ppark.groupby(['PARENT_NAME']).sum()
park_group
```

For interpreting this table, we need to measure some of the central tendency variables.

```python
medianpg = park_group['ACREAGE'].median()
meanpg = park_group['ACREAGE'].mean()
print('Median value:', medianpg)
print('Mean value:', meanpg)
```

```python
ax = park_group.ACREAGE.hist(bins=35)
ax.axvline(meanpg, color='red', linestyle='solid', linewidth=1.6, alpha = .8)
ax.axvline(medianpg, color='green', linestyle='solid', linewidth=1.6, alpha = .8)
ax.legend([meanpg, medianpg], labels = ['mean', 'median'])
ax.set_xlabel('Distribution of sizes of the parent parks [ac]')
```

Figure 3: This is the distribution of Philadelphia's park grouped by the parent name of the park. This type of grouping increases the park size and consequently the mean value because small parks with less area are part of a bigger group, and their areas are added together.

```python
stdpg = park_group['ACREAGE'].std()
print('stdpg:', stdpg)
```

Standard deviation of the data grouped under Parent_name is greater that the original dataset, showing the distribution is wider in grouping by Parent_name.
